# Runge Kutta 2nd Order Method
## Example
Find $y(0.2)$ for the differential equation $\frac{dy}{dx} = -xy, y(0) = 1$ using Runge Kutta 2nd order method, taking $h = 0.02$
# (a) Heun's Method
## $k_{1} = hf(x_{n}, y_{n})$
## $k_{2} = hf(x_{n} + h, y_{n} + k_{1})$
## $y_{n+1} = y_{n} + \frac{1}{2}(k_{1} + k_{2})$ 

```python
import math
import matplotlib.pyplot as plt
```

```python
#Define the Functions

#The differential equation is dy/dy = -xy with initial conditions y(0)=1
dy = lambda x,y: -x*y

#'f' represent the exact solution of the differential equation
f = lambda x: math.exp(-x**2/2)  
```

```python
#Initialize Values

#where 'x0 = 0' represents the initial value of x
x0 = 0

#'xf = 0.2' represents the final value of x
xf = 0.2

#'h = 0.02' represents the step size
h = 0.02

#'n' represents the number of steps
n = int((xf - x0)/h)

#'x0 = 0' and 'y0 = 1' represent the initial condition of x and y
x0 = 0; y0 = 1

x = x0
y = y0

#Visualization code for initialization

#Header for the table
print('x \t\t y \t\t f(x)')

#The initial values of x, y and f(x)
print('%f \t %f \t %f' %(x,y,f(x)))

#Initialize list for plotting
x_plot = []; y_heuns = []; y_analytical = []

#Heun's method implementation
for i in range (1, n+1):
    x_plot.append(x)
    y_heuns.append(y)
    y_analytical.append(f(x))
    
    #Calculating the differential equation using Heun's method
    k1 = h * dy(x,y)
    k2 = h * dy(x + h, y + k1)
    #Approximate solution of y
    y = y + 1/2 * (k1 + k2)
    
    #Increase x by step size to calculate the next y estimate
    x = x + h
    
    print('%f \t %f \t %f' %(x, y, f(x)))
    
#Visualization codes for plotting the approximate and analytical solution of the differential equation
x_plot.append(x)
y_heuns.append(y)
y_analytical.append(f(x))
plt.plot(x_plot,y_heuns, 'ro--', x_plot, y_analytical)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(["Heun's", 'Analytical'])
plt.title("Heun's Method", fontsize=20)
plt.grid(True)
```

# (a) Mid-point Method
## $k_{1} = hf(x_{n}, y_{n})$
## $k_{2} = hf(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{1}}{2})$
## $y_{n+1} = y_{n} + k_{2}$ 

```python
#Initialize Values

#where 'x0 = 0' represents the initial value of x
x0 = 0

#'xf = 0.2' represents the final value of x
xf = 0.2

#'h = 0.02' represents the step size
h = 0.02

#'n' represents the number of steps
n = int((xf - x0)/h)

#'x0 = 0' and 'y0 = 1' represent the initial condition of x and y
x0 = 0; y0 = 1

x = x0
y = y0

#Visualization code for initialization

#Header for the table
print('x \t\t y \t\t f(x)')

#The initial values of x, y and f(x)
print('%f \t %f \t %f' %(x,y,f(x)))

#Initialize list for plotting
x_plot = []; y_midpoint = []; y_analytical = []

#Mid point method implementation
for i in range (1, n+1):
    x_plot.append(x)
    y_midpoint.append(y)
    y_analytical.append(f(x))
    
    #Calculating the differential equation using Mid point method
    k1 = h * dy(x,y)
    k2 = h * dy(x + h/2, y + k1/2)
    #Approximate solution of y
    y = y + k2
    
    #Increase x by step size to calculate the next y estimate
    x = x + h
    
    print('%f \t %f \t %f' %(x, y, f(x)))
    
#Visualization codes for plotting the approximate and analytical solution of the differential equation
x_plot.append(x)
y_midpoint.append(y)
y_analytical.append(f(x))
plt.plot(x_plot,y_midpoint, 'ro--', x_plot, y_analytical)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(["Mid point", 'Analytical'])
plt.title('Mid point Method', fontsize=20)
```

# (a) Ralston's Method
## $k_{1} = hf(x_{n}, y_{n})$
## $k_{2} = hf(x_{n} + \frac{3}{4}h, y_{n} + \frac{3}{4}k_{1})$
## $y_{n+1} = y_{n} + \frac{1}{3}(k_{1} + 2k_{2})$ 

```python
#Initialize Values

#where 'x0 = 0' represents the initial value of x
x0 = 0

#'xf = 0.2' represents the final value of x
xf = 0.2

#'h = 0.02' represents the step size
h = 0.02

#'n' represents the number of steps
n = int((xf - x0)/h)

#'x0 = 0' and 'y0 = 1' represent the initial condition of x and y
x0 = 0; y0 = 1

x = x0
y = y0

#Visualization code for initialization

#Header for the table
print('x \t\t y \t\t f(x)')

#The initial values of x, y and f(x)
print('%f \t %f \t %f' %(x,y,f(x)))

#Initialize list for plotting
x_plot = []; y_ralstons = []; y_analytical = []

#Ralston's method implementation
for i in range (1, n+1):
    x_plot.append(x)
    y_ralstons.append(y)
    y_analytical.append(f(x))
    
    #Calculating the differential equation using Ralston's method
    k1 = h * dy(x,y)
    k2 = h * dy(x + 3*h/4, y + 3*k1/4)
    #Approximate solution of y
    y = y + 1/3 * (k1 + 2*k2)
    
    #Increase x by step size to calculate the next y estimate
    x = x + h
    
    print('%f \t %f \t %f' %(x, y, f(x)))
    
#Visualization codes for plotting the approximate and analytical solution of the differential equation
x_plot.append(x)
y_ralstons.append(y)
y_analytical.append(f(x))
plt.plot(x_plot,y_ralstons, 'ro--', x_plot, y_analytical)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(["Ralston's", 'Analytical'])
plt.title("Ralston's Method", fontsize=20)
plt.grid(True)
```
