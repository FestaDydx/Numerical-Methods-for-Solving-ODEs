# Runge Kutta 4th Order Method
## $k_{1} = hf(x_{n}, y_{n})$
## $k_{2} = hf(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{1}}{2})$
## $k_{3} = hf(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{2}}{2})$
## $k_{4} = hf(x_{n} + h, y_{n} + k_{3})$
## $y_{n+1} = y_{n} + \frac{1}{6}(k_{1} + 2k_{2} + 2k_{3} + k_{4})$ 
## Example
Apply the Runge Kutta 4th Method to obtain $y(5)$ from the differential equation $\frac{dy}{dx} = xy^\frac{1}{3}, y(1) = 1$ taking $h = 0.1$

```python
import math
import matplotlib.pyplot as plt
```

```python
#Define the Functions

#The differential equation is dy/dy = xy^1/3 with initial conditions y(1)=1
dy = lambda x,y: x*y**(1/3)

#'f' represent the exact solution of the differential equation
f = lambda x: (math.sqrt((x**2 + 2)/3))**3    
```

```python
#Initialize Values

#where 'x0 = 1' represents the initial value of x
x0 = 1

#'xf = 5' represents the final value of x
xf = 5

#'h = 0.1' represents the step size
h = 0.1

#'n' represents the number of steps
n = int((xf - x0)/h)

#'x0 = 0' and 'y0 = 2' represent the initial condition of x and y
x0 = 1; y0 = 1

x = x0
y = y0

#Visualization code for initialization

#Header for the table
print('x \t\t y \t\t f(x)')

#The initial values of x, y and f(x)
print('%f \t %f \t %f' %(x,y,f(x)))

#Initialize list for plotting
x_plot = []; y_RK4 = []; y_analytical = []

#Runge Kutta 4th order method implementation
for i in range (1, n+1):
    x_plot.append(x)
    y_RK4.append(y)
    y_analytical.append(f(x))
    
    #Calculating the differential equation using RK4 method
    k1 = h * dy(x,y)
    k2 = h * dy(x + h/2, y + k1/2)
    k3 = h * dy(x + h/2, y + k2/2)
    k4 = h * dy(x + h, y + k3)
    #Approximate solution of y
    y = y + 1/6 * (k1 + 2*k2 + 2*k3 + k4)
    
    #Increase x by step size to calculate the next y estimate
    x = x + h
    
    print('%f \t %f \t %f' %(x, y, f(x)))
    
#Visualization codes for plotting the approximate and analytical solution of the differential equation
x_plot.append(x)
y_RK4.append(y)
y_analytical.append(f(x))
plt.plot(x_plot,y_RK4, 'ro--', x_plot, y_analytical)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(["RK", 'Analytical'])
plt.title('Runge-Kutta 4th Order Method', fontsize=20)
```
