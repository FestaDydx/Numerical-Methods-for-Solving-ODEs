# Euler's Method
# $y_{n+1} = y_{n} + hf(x_{n}, y_{n})$

## Example
Use the Euler's Method to solve the IVP $\frac{dy}{dx} = -20y, y(0) = 1.$ Compute the values of $y(1)$ with the stepsize $h = 0.02$

```python
import math
import matplotlib.pyplot as plt
```

```python
#Define the Functions

#The differential equation is dy/dx = -20y with initial conditions y(0)=1
dy = lambda x,y: -20*y

#'f' represent the exact solution of the differential equation
f = lambda x: math.exp(-20*x)  
```

```python
#Initialize Values

#where 'x0 = 0' represents the initial value of x
x0 = 0

#'xf = 1' represents the final value of x
xf = 1

#'h = 0.02' represents the step size
h = 0.02

#'n' represents the number of steps
n = int((xf - x0)/h)

#'x0 = 0' and 'y0 = 1' represent the initial condition of x and y
x0 = 0; y0 = 1

x = x0
y = y0

#Visualization code for initialization

#Header for the table
print('x \t\t y \t\t f(x)')

#The initial values of x, y and f(x)
print('%f \t %f \t %f' %(x,y,f(x)))

#Initialize list for plotting
x_plot = []; y_euler = []; y_analytical = []

#Euler method implementation
for i in range (1, n+1):
    x_plot.append(x)
    y_euler.append(y)
    y_analytical.append(f(x))
    
    #Calculating the differential equation using Euler method
    y = y + h * dy(x,y)
    
    #Increase x by step size to calculate the next y estimate
    x = x + h
    
    print('%f \t %f \t %f' %(x, y, f(x)))
    
#Visualization codes for plotting the approximate and analytical solution of the differential equation
x_plot.append(x)
y_euler.append(y)
y_analytical.append(f(x))
plt.plot(x_plot,y_euler, 'ro--', x_plot, y_analytical)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(["Euler", 'Analytical'])
plt.title("Euler's Method", fontsize=20)
plt.grid(True)
```
